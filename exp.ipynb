```python
df_copy = df.copy()
```

```python
df_copy.tail()
```

```python
df_copy['Roles'].str.split('.')
```

```python
df_copy['Roles'].loc[0]
```

```python
df_copy['Roles'].str.split('•', expand = True)
```

```python
# unique roles
df_copy["Roles"].str.split('•').explode().str.strip().unique()
```

```python
df_copy["Roles"].str.split('•').explode().str.strip().value_counts()
```

```python
df_copy[['role1', 'role2', 'role3', 'role4', 'role5',]] = df_copy['Roles'].str.split('•', expand = True)
df_copy["Roles"].loc[675]
```

```python
# split name into first and second name
df_copy["Full name"]
```

```python
# split into date and born place
df_copy["Born"]
date_pattern = r"(\d+ \w+ \d{4})"
df_copy["date born"] = df_copy["Born"].str.extract(date_pattern)
df_copy["date died"] = df_copy["Died"].str.extract(date_pattern)

df_copy.head(2)
```

```python
# split into country
country_pattern = r"(\(\w+\))"
df_copy["country_born"] = df_copy["Born"].str.extract(country_pattern)
df_copy.head(2)
```

```python
# split into height and weight
df_copy[["height_cm", "weight_kg"]] = df_copy["Measurements"].str.split("/", expand = True)

# which country has the tallest competitor on average
df_copy[["height_cm", "weight_kg"]]
```

```python
height = df_copy[["NOC", "height_cm"]]
height["height_cm"] = height["height_cm"].str.extract(r"(\d+)").astype(float)
height.groupby("NOC")["height_cm"].mean().sort_values(ascending = False)
```

```python
results = pd.read_csv("results.csv")
results.head(2)
```

```python
results["Medal"].unique()
```

```python
results.groupby(["NOC", "Medal"]).size().unstack(fill_value = 0).sort_values("Gold", ascending = False)
```

```python
gold_streaks = results[results["Medal"] == "Gold"].groupby(["athlete_id", "NOC", "Games"]).size().to_frame(name="Gold Streak").reset_index("Gold Streak")
silver_streaks = results[results["Medal"] == "silver"].groupby(["athlete_id", "NOC", "Games"]).size().to_frame(name="Silver Streak").reset_index("Silver Streak")
bronze_streaks = results[results["Medal"] == "Bronze"].groupby(["athlete_id", "NOC", "Games"]).size().to_frame(name="Bronze Streak").reset_index("Broze Streak")
```
